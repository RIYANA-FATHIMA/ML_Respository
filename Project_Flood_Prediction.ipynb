```python


#capping method
def iqr_capping(df,variables):
  q1=df[variables].quantile(0.25)
  q3=df[variables].quantile(0.75)
  iqr=q3-q1
  upper_whisker=q3+(1.5*iqr)
  lower_whisker=q1-(1.5*iqr)
  return lower_whisker,upper_whisker
```

```python
lower_1,upper_1=iqr_capping(data,'MAR')
print('lower 1= ',lower_1)
print('upper 1= ',upper_1)
data['JAN']=np.where(data['JAN']>upper_1,upper_1,np.where(data['JAN']<lower_1,lower_1,data['JAN']))
```

# **Modelling**

```python
y=data['flood']
x=data.drop(['flood'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

**logistic regression**

```python

from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred=logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score
print('Accuracy is ',accuracy_score(y_test,y_pred))
print('Precision is ',precision_score(y_test,y_pred))
print('Recall is ',recall_score(y_test,y_pred))
print('f1 is ',f1_score(y_test,y_pred))
confusion_matrix(y_test,y_pred)
```

**KNN**

```python

from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,15)

for k in neighbors:
  classifier = KNeighborsClassifier(n_neighbors=k,metric='minkowski',p=1)
  classifier.fit(x_train,y_train)
  y_pred = classifier.predict(x_test)
  acc = accuracy_score(y_test,y_pred)
  metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors=10,metric='minkowski',p=1)
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
y_pred
```

```python
print('Accuracy is', accuracy_score(y_test,y_pred))
print('Precision is', precision_score(y_test,y_pred))
print('Recall is', recall_score(y_test,y_pred))
print('f1 score is', f1_score(y_test,y_pred))
confusion_matrix(y_test,y_pred) 
```

**SVM**

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy is: ', accuracy_score(y_test,y_pred))
```

**Decision Tree**

```python
print('Accuracy is: ',accuracy_score(y_test,y_pred))
```
